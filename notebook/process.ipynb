```python
df_weather['date'] = pd.to_datetime(df_weather['date'])
df['date'] = pd.to_datetime(df['date'])
df_merged = pd.merge(df, df_weather, on='date', how='inner')
```

```python
df_merged.drop(columns=['Month','NGÀY','date_time'])
```

```python
df=df_merged
df = df[df['D THU'] != 0]
df['revenue']=df['D THU']
```

```python

# 1. Doanh thu hôm trước
df['revenue_prev1'] = df['revenue'].shift(1)

# 2. Doanh thu 2 ngày trước
df['revenue_prev2'] = df['revenue'].shift(2)

# 3. Trung bình doanh thu 3 ngày trước
df['revenue_mean_3'] = df['revenue'].shift(1).rolling(window=3).mean()

# 4. Trung bình doanh thu 7 ngày trước
df['revenue_mean_7'] = df['revenue'].shift(1).rolling(window=7).mean()

# 5. Độ chênh lệch doanh thu giữa hôm trước và hôm trước nữa
df['revenue_diff'] = df['revenue_prev1'] - df['revenue_prev2']

# 6. Tăng trưởng phần trăm giữa hôm trước và hôm trước nữa
df['revenue_pct_change'] = (df['revenue_prev1'] - df['revenue_prev2']) / df['revenue_prev2']

# 7. Độ lệch so với trung bình 7 ngày gần nhất
df['revenue_dev_from_mean7'] = df['revenue_prev1'] - df['revenue_mean_7']
```

```python
df=df.drop(columns=['Month','NGÀY','date_time'])
```
